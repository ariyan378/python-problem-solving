```python
import json

class Person:
    def __init__(self, name, id):
        self.name = name 
        self.id = id

person = Person('Hridoy', 378)

def display_object(obj): # Changed parameter name to 'obj'
    if isinstance(obj, Person):
        # Use 'obj.name', NOT 'Person.name'
        return "My name is {} . My Id is {}".format(obj.name, obj.id)

with open('Object.json', 'w') as f:
    # Use 'person' (your variable), NOT 'object' (the built-in type)
    json.dump(person, f, default=display_object) 
         
with open('Object.json', 'r') as f:
    text = json.load(f)
    print(text)
```
