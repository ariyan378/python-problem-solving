````python
#python and numpy speedtest

#pythonic way
import numpy as np
import time

size = 1000000

py_data = list(range(size))
py_start = time.time()
py_result = [ x * 1.10 for x in py_data]
py_time = time.time()- py_start


#numpy way

np_array = np.arange(size, dtype = float)
np_start = time.time()
np_result = np_array * 1.10
np_time = time.time()- np_start


print(f" Python time {py_time:.2f}")
print(f" Numpy time {np_time:.2f}")
print(f" Numpy faster {(py_time/np_time):.1f}x Times")
````

### Section 4 — Indexing & Slicing

#### 4.1 The Mental Model

Think of indexing like **navigating a spreadsheet**:
- `array[row, column]` — pick one cell
- `array[start:stop]` — pick a range (stop is EXCLUDED)
- `array[::step]` — pick every nth element

#### 4.2 1D Indexing

````python
temp = np.arange(16)
temp
````

````python
temp[1:5:2]
````

````python
temp = np.arange(16)
temp
````

````python
# Single element access
print("Day 1 temp (index 0) :", temp[0])    # 22
print("Day 7 temp (index 6) :", temp[6])    # 40
print("Last day (index -1)  :", temp[-1])   # 30  ← negative indexing!
print("2nd to last (index-2):", temp[-2])   # 23
print()

# Slicing: array[start : stop : step]  (stop is NOT included)
print("First 3 days:", temp[0:3])     # [22 19 35]
print("Days 4-7:     ", temp[3:7])    # [28 31 17 40]
print("Last 3 days:  ", temp[-3:])    # [26 23 30]
print("Every 2nd day:", temp[::2])    # [22 35 31 40 23]
print("Reversed:      ", temp[::-1])
````

#### Section 5 — Boolean Indexing & Filtering

#### 5.1 Why Boolean Indexing Is Your Best Tool

In Excel, you use **filters** to show only rows matching a condition. In NumPy, that's boolean indexing — it's faster and more powerful.

The idea:
1. Create a **boolean mask** (array of True/False)
2. Use that mask to **select elements** where True

````python
daily_sales = np.array([320, 450, 380, 290, 510, 620, 480])
target = 350

print("sales more than 400 :" ,daily_sales>target )
````

````python
target_done = daily_sales>350
print(target_done)
target_done.sum()
````

````python
# & and | or ~ not

ages = np.array([34, 67, 23, 45, 78, 56, 12, 89, 43, 61])


working_age = (ages >= 30) & (ages <= 65)
print("Working age patients:", ages[working_age])
````
